### 6. Join the dataframes

```python
# Join 2 dataframes using as key the NOC

df_merged = pd.merge(df_athlete_events, df_regions, on='NOC', how='left')
df_merged.head()
```

```python
# Basic information about df_merged

df_merged.info()
```

```python
# Check if the dataframe contains NaN values

df_merged.isnull().any()
```

```python
# Total number and percentage of missing data per column

print(f"Number of NaN values out of {df_merged.shape[0]} rows per column")
total = df_merged.isnull().sum().sort_values(ascending=False)
percent = (df_merged.isnull().sum()/df_merged.isnull().count() * 100).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data
```

#### We can see that 6 columns contain NaN values. 
#### Medal column has 85% of missing values.

#### To make it easier to analyze France, I created a new dataframe.

```python
# Create dataframe with only French athletes

df_france = df_merged[(df_merged.region == 'France')]
df_france.reset_index()
```

```python
# Basic information about df_france

df_france.info()
```

```python
# Total number and percentage of missing data per column

print(f"Number of NaN values out of {df_france.shape[0]} rows per column")
total = df_france.isnull().sum().sort_values(ascending=False)
percent = (df_france.isnull().sum()/df_france.isnull().count() * 100).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data
```

### 7. Years in which Olympic Games were held

```python
# Show years of Summer Olympic Games

summer_olympics = np.array(sorted(df_merged[df_merged['Season']=='Summer']['Year'].unique()))
print('Summer Olypics were held in the years:', summer_olympics)
```

```python
# Show years of Winter Olympic Games

winter_olympics = np.array(sorted(df_merged[df_merged['Season']=='Winter']['Year'].unique()))
print('Winter Olypics were held in the years:', winter_olympics)
```

### 8. Cities that have hosted Olympic Games

```python
# Cities that have hosted Games

cities = df_merged.groupby('City').apply(lambda x:x['Year'].unique()).to_frame().reset_index()
cities.columns = ['City', 'Years']
cities['Occurrence'] = [len(c) for c in cities['Years']]
cities.sort_values('Occurrence', ascending=False).reset_index(drop=True)
```

### We can see that 42 cities have hosted the Olympic Games.
### 4 French cities are in the table above.
### PARIS was 2 times city host of Summer Games (1900, 1924).
### ALBERTVILLE, CHAMONIX, GRENOBLE were 1 time city hosts of Winter Games (1992, 1924, 1968).

### 9. Participation of Men and Women at the Olympic Games

```python
# Show total number of athletes in the Olympic Games

total_num_athlets = len(df_merged.ID.unique())
print('Total number of athletes in the Olympic Games is', total_num_athlets)
```

```python
# Show total number of women

num_women = df_merged[(df_merged.Sex == 'F') & (df_merged.Season)]
total_num_women = len(num_women.ID.unique())
total_num_women                                                
```

```python
# Show total number of men

num_men = df_merged[(df_merged.Sex == 'M') & (df_merged.Season)]
total_num_men = len(num_men.ID.unique())
total_num_men   
```

#### For the last 120 years, presence of male athletes at the Olympic Games has been almost 3 TIMES GREATER than the presence of female athletes.

```python
# Participation of men and women at the Olympic Games

df_merged.groupby('Year')['Sex'].value_counts()
```

```python
# Plot participation of men and women at the Olympic Games

part = df_merged.groupby('Year')['Sex'].value_counts()

part.loc[:,'F'].plot(label='Women', figsize=(13, 5), title='Men/Women participation at the OG')
part.loc[:,'M'].plot(label='Men')

plt.ylabel("Athletes")
plt.legend(loc='upper left')
plt.show()
```

#### Drastic domination of men was until 1990s.

### 12. Medals per country

```python
# Total number of medals per country

medals_country = df_merged.groupby('region')['Medal'].value_counts().reset_index(name='Total')
medals_country
```

```python
# Top ten nations by gold medals won

gold = df_merged[df_merged.Medal == "Gold"].region.value_counts().reset_index(name='Gold Medal')
gold.head(10)
```

```python
# Top ten nations by silver medals won

silver = df_merged[df_merged.Medal == "Silver"].region.value_counts().reset_index(name='Silver Medal')
silver.head(10)
```

```python
# Top ten nations by bronze medals won

bronze = df_merged[df_merged.Medal == "Bronze"].region.value_counts().reset_index(name='Bronze Medal')
bronze.head(10)
```

```python
# Top ten nations by medals won

plt.subplot(3,1,1)
gold = df_merged[df_merged.Medal == "Gold"].region.value_counts().head(10)
gold.plot(kind='bar', rot=0, figsize=(15, 10), color=['gold'])
plt.ylabel("Gold Medal")

plt.subplot(3,1,2)
silver = df_merged[df_merged.Medal == "Silver"].region.value_counts().head(10)
silver.plot(kind='bar', rot=0, figsize=(15, 10), color=['silver'])
plt.ylabel("Silver Medal")

plt.subplot(3,1,3)
bronze = df_merged[df_merged.Medal == "Bronze"].region.value_counts().head(10)
bronze.plot(kind='bar', rot=0, figsize=(15, 10), color=['brown'])
plt.ylabel("Bronze Medal")

plt.show()
```

### USA is the country with the most medals won (gold = 2638, silver = 1641, bronze = 1358).

### France is on the 6th place in the world with 501 gold medals.
### France is on the 5th place in the world with 610 silver medals.
### France is on the 4th place in the world with 666 bronze medals.
